# Comparison of Asset Classes

# Stock Mean

```python
#Stock Close - Preston rename close df to 
stock_ticker_close_df
```

```python
#Bond Close - Joseph rename close df to bond_ticker_close_df
stocks_daily_returns = stock_ticker_close_df.pct_change()
stocks_daily_returns.dropna(inplace = True)
stocks_daily_returns
```

```python
stocks_daily_returns_mean = stocks_daily_returns.mean(axis=1)
```

```python
stocks_daily_returns_mean
```

```python
stocks_daily_returns_mean.hvplot(label = 'Mean Daily Returns of Stocks', ylabel = 'Mean Daily Return %', height = 500, width = 1000, color = 'r')
```

```python
stocks_cumulative_returns_mean = (1 + stocks_daily_returns_mean).cumprod() - 1
stocks_cumulative_returns_mean.hvplot(label = 'Mean Cumulative Returns of Stocks', ylabel = 'Mean Cumulative Return %', height = 500, width = 1000, color = 'r')
```
